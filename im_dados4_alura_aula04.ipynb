### Utilizando gráficos

```python
# vamos descobrir quais são os 10 bairros que tem mais anúncios
dados_vendas['Bairro'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# inserimos essa query dentro de ‘data’
plt.figure(figsize=(16, 8))
ax = sns.barplot(x='Bairro', y='Valor_M2_bairro', data=dados_bairro.query('Bairro in @top_bairro'))
ax.set_title('Valor do m2 dos 10 bairros com mais anúncios de SP')
ax.tick_params(axis='x', rotation=45)
```

```python
# plotando um boxplot, podemos ver que temos um outlier dentro do Jardim Europa

plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro'), x='Bairro', y='Valor_anuncio_float')
plt.show()
```

```python
# plotando o mesmo gráfico, só que analisando a metragem, vemos que temos um
# imóvel muito diferente dentro do Jardim Guedala
plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro'), x='Bairro', y='Metragem')
plt.show()
```

```python
# podemos removê-lo colocando uma metragem máxima permitida na query
plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro & Metragem < 30000'), x='Bairro', y='Metragem')
plt.show()
```
