# Come√ßando com os desafios passados... 

```python
dados_vendas.head()
```

```python
# c√°lculo do valor pelo M2
dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
```

```python
# cria√ß√£o da coluna Valor_M2 dentro da df
dados_vendas['Valor_M2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
dados_vendas.head()
```

```python
# agrupando por bairro e vendo a m√©dia
dados_vendas.groupby('Bairro').mean()
```

```python
# criando um novo df (dados_bairro) e agrupando por bairro com a soma
dados_bairro = dados_vendas.groupby('Bairro').sum()
dados_bairro
```

```python
# criando a coluna Valor_M2_bairro utilizando a df agrupada por bairro e soma
dados_bairro['Valor_M2_bairro'] = dados_bairro['Valor_anuncio_float']/dados_bairro['Metragem']
dados_bairro
```

```python
# agrupando a df dados_vendas por bairro e m√©dia novamente para compara√ß√£o
dados_vendas.groupby('Bairro').mean()
```

```python
# vamos fazer agora com a Vila Mariana usando a 'm√©dia de m√©dias'
dados_vendas.groupby('Bairro').mean().loc['Vila Mariana']
```

```python
# comparando com o valor obtido usando a soma
dados_bairro.loc['Vila Mariana']
```

```python
dados_bairro.loc['Vila Mariana']['Valor_M2_bairro']
```

```python
dados_bairro['Valor_M2_bairro']['Vila Mariana']
```

## Utilizando gr√°ficos

```python
# vamos descobrir quais s√£o os 10 bairros que tem mais an√∫ncios
dados_vendas['Bairro'].value_counts()
```

```python
top_bairro = dados_vendas['Bairro'].value_counts()[:10].index
```

```python
# precisamos do bairro n√£o como √≠ndice, mas como coluna
dados_bairro
```

```python
# atribuindo o par√¢metro inplace para que n√£o seja necess√°rio atribuir 
# as mudan√ßas a uma nova df
dados_bairro.reset_index(inplace=True)
```

```python
dados_bairro.query('Bairro in @top_bairro')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# inserimos essa query dentro de ‚Äòdata‚Äô
plt.figure(figsize=(16, 8))
ax = sns.barplot(x='Bairro', y='Valor_M2_bairro', data=dados_bairro.query('Bairro in @top_bairro'))
ax.set_title('Valor do m2 dos 10 bairros com mais an√∫ncios de SP')
ax.tick_params(axis='x', rotation=45)
```

```python
# plotando um boxplot, podemos ver que temos um outlier dentro do Jardim Europa

plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro'), x='Bairro', y='Valor_anuncio_float')
plt.show()
```

```python
# plotando o mesmo gr√°fico, s√≥ que analisando a metragem, vemos que temos um
# im√≥vel muito diferente dentro do Jardim Guedala
plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro'), x='Bairro', y='Metragem')
plt.show()
```

```python
# podemos remov√™-lo colocando uma metragem m√°xima permitida na query
plt.figure(figsize=(16, 8))
ax=sns.boxplot(data=dados_vendas.query('Bairro in @top_bairro & Metragem < 30000'), x='Bairro', y='Metragem')
plt.show()
```
