# Desafio \#01 ‚úÖ
Realizar a m√©dia da metragem para cada um dos bairros. (Proposto pelo Paulo)

```python
# criando uma df (bairro_metragem) apenas com os Bairros e Metragem de cada item
bairro_metragem = dados.filter(items=['Bairro', 'Metragem'])
```

```python
# agrupando linhas iguais com base nos dados da coluna 'Bairro'
bairro_metragem.groupby('Bairro')
# calculando a m√©dia correspondente na coluna 'Metragem'
md_bairro_metragem = bairro_metragem.groupby('Bairro').mean().round(2)
md_bairro_metragem
```

# Desafio \#05 ‚úÖ
Descobrir quais s√£o os bairros que n√£o tem nome de rua. (Proposto pela Vivian)
