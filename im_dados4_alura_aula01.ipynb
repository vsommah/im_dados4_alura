# Desafio \#01 ✅
Realizar a média da metragem para cada um dos bairros. (Proposto pelo Paulo)

```python
# criando uma df (bairro_metragem) apenas com os Bairros e Metragem de cada item
bairro_metragem = dados.filter(items=['Bairro', 'Metragem'])
```

```python
# agrupando linhas iguais com base nos dados da coluna 'Bairro'
bairro_metragem.groupby('Bairro')
# calculando a média correspondente na coluna 'Metragem'
md_bairro_metragem = bairro_metragem.groupby('Bairro').mean().round(2)
md_bairro_metragem
```

# Desafio \#02 ✅
Duas formas de selecionar os dados por bairro (consultar os métodos na documentação do Pandas). (Proposto pelo Thiago)

```python
# utilizando uma df criada durante a Aula 01 para criar uma seleção de todos
# imóveis com mais de 3 banheiros dentro da Vila Mariana em ordem descendente
imoveis_vila_mariana.loc[imoveis_vila_mariana['Banheiros'] > 3].sort_values(by='Banheiros', ascending=False)
```

```python
# praticando localização com duas condições
dados.loc[(dados['Bairro'] == 'Planalto Paulista') & (dados['Banheiros'] > 3)].sort_values(by='Banheiros', ascending=False)
```

# Desafio \#03 ✅
Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF (dataframe), assim como tentar colocar alguma conclusão.

```python
# gráfico de barras utilizando df de bairro x metragem média
# apresentaçaõ dos 10 últimos valores de metragem média em ordem decrescente
plot_b_m = md_bairro_metragem.sort_values('Metragem', ascending=False)
plot_b_m.tail(10).plot.bar()
```

# Desafio \#05 ✅
Descobrir quais são os bairros que não tem nome de rua. (Proposto pela Vivian)

```python
# criando um novo df com a filtragem de Bairros x Ruas
bairro_ruas = dados.filter(items=['Bairro', 'Rua'])
bairro_ruas
```

```python
# utilizando o df anterior para verificar quais items não possuem nome da Rua
bairro_sem_ruas = bairro_ruas[bairro_ruas['Rua'].isnull()]
bairro_sem_ruas = bairro_sem_ruas['Bairro'].value_counts()
bairro_sem_ruas
```
